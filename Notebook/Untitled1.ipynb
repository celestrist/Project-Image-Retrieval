```python
from scipy.cluster.vq import vq, kmeans
```

```python
img = array(imread("/Users/Celestrist/Pictures/833297815.jpeg"))
imshow(img,origin="lower")
```

```python
img_array = img.reshape(img.shape[0]*img.shape[1],3)
centroid,_ = kmeans(img_array,10)
index,_ = vq(img_array,centroid)
quantized = centroid[index].reshape(img.shape[0],img.shape[1],3)
imshow(quantized,origin="lower")
```

```python
from scipy.stats.distributions import entropy
```

```python
maximum_entropy = entropy(ones(img_array.shape))
minimum_entropy = entropy(random.random(img_array.shape))
original_entropy = entropy(img_array)
quantized_entropy = entropy(quantized.reshape(-1,3))

def scale(x):
    return (x - minimum_entropy)/(maximum_entropy-minimum_entropy)

print scale(original_entropy),scale(quantized_entropy)
    
    
```
